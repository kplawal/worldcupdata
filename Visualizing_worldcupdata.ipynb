```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WorldCupMatches.csv')
print(df.head())
```

```python
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
print(df['Total Goals'])
```

```python
# A bar chart visualizing showing the average amount of goals scored in a match each year during the World Cup between 1930-2014.
sns.set_style('whitegrid')
sns.set_context('poster',font_scale = 0.8)
fig, ax = plt.subplots(figsize = (20,7))
ax = sns.barplot(data = df,
           x = 'Year',
           y = 'Total Goals')
ax.set_title('Average Number Of Goals Scored In World Cup Matches By Year')
plt.show()
```
